```python
import os

import matplotlib
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point

from pykrige.ok import OrdinaryKriging

from air_brain.data.get_data import DATA_DIR
from air_brain.util.air import PM25
from air_brain.util.od import od

bg_file = os.path.join(DATA_DIR, "tl_2010_42003_bg10", "tl_2010_42003_bg10.shp")
tract_file = os.path.join(DATA_DIR, "tl_2010_42003_tract10", "tl_2010_42003_tract10.shp")
zipcode_file = os.path.join(DATA_DIR, "zipcodes.geojson")
county_file = os.path.join(DATA_DIR, "county.geojson")
sensor_file = os.path.join(DATA_DIR, "sensor_json.geojson")
```

```python
bgs = gpd.read_file(bg_file)
bgs["ID"] = bgs.GEOID10.astype(int)
```

```python
tracts = gpd.read_file(tract_file)
tracts["ID"] = tracts.GEOID10.astype(int)
```

```python
zipcodes = gpd.read_file(zipcode_file)
zipcodes.ZIP = zipcodes.ZIP.astype(int)
```

```python
def plot(year):
    fig, axs = plt.subplots(2, 3)
    # PM 2.5
    # block group
    aq_bg_file = os.path.join(DATA_DIR, "epa_ej", "{}.csv".format(year))
    aq_bg = pd.read_csv(aq_bg_file)
    df = bgs.merge(aq_bg, on="ID", how="right", validate="1:1")
    df.plot(column="PM25", cmap="Reds", legend=True, ax=axs[0,0])
    # tract
    aq_tract_file = os.path.join(DATA_DIR, "epa_ej", "{}_tract.csv".format(year))
    aq_tract = pd.read_csv(aq_tract_file)
    df = tracts.merge(aq_tract, on="ID", how="right", validate="1:1")
    df.plot(column="PM25", cmap="Reds", legend=True, ax=axs[0,1])
    # zipcode
    aq_zip_file = os.path.join(DATA_DIR, "epa_ej", "{}_zipcode.csv".format(year))
    aq_zip = pd.read_csv(aq_zip_file)
    df = zipcodes.merge(aq_zip, on="ZIP", how="right", validate="1:1")
    df.plot(column="PM25", cmap="Reds", legend=True, ax=axs[0,2])
    # ozone
    # block group
    aq_bg_file = os.path.join(DATA_DIR, "epa_ej", "{}.csv".format(year))
    aq_bg = pd.read_csv(aq_bg_file)
    df = bgs.merge(aq_bg, on="ID", how="right", validate="1:1")
    df.plot(column="O3", cmap="Blues", legend=True, ax=axs[1,0])
    # tract
    aq_tract_file = os.path.join(DATA_DIR, "epa_ej", "{}_tract.csv".format(year))
    aq_tract = pd.read_csv(aq_tract_file)
    df = tracts.merge(aq_tract, on="ID", how="right", validate="1:1")
    df.plot(column="O3", cmap="Blues", legend=True, ax=axs[1,1])
    # zipcode
    aq_zip_file = os.path.join(DATA_DIR, "epa_ej", "{}_zipcode.csv".format(year))
    aq_zip = pd.read_csv(aq_zip_file)
    df = zipcodes.merge(aq_zip, on="ZIP", how="right", validate="1:1")
    df.plot(column="O3", cmap="Blues", legend=True, ax=axs[1,2])

    return fig, axs
```
